##### Nova DF amb una selecció de les columnes

```python
airdf=df.iloc[:, [1,3,4,6,8,11,13,14,15,16,17,18,21,23,24,25,26,27,28]]
```

```python
airdf.head(3)
```

##### nova columna, status, que es pot fer servir després com a target

```python
airdf.loc[airdf['ArrDelay'] < 15, 'status'] = 'ontime'
airdf.loc[airdf['ArrDelay'] >= 15, 'status'] = 'delayed'
airdf.loc[airdf['Cancelled'] == 1, 'status'] = 'cancelled'
airdf.loc[airdf['Diverted'] == 1, 'status'] = 'diverted'
```

```python
airdf.isnull().sum()
```

```python
#proporció de vols retrassats, puntuals, cancel.lats i desviats
airdf['status'].value_counts(normalize=True) * 100
airdf
```

```python
#eliminem les columnes
airdf=airdf.drop(['Cancelled'], axis=1)
airdf=airdf.drop(['Diverted'], axis=1)
```

```python
#eliminem les observacions amb cancel.lacions o desviaments
airdf.drop(airdf[airdf['status'] == 'diverted'].index, inplace = True)
airdf.drop(airdf[airdf['status'] == 'cancelled'].index, inplace = True)
```

```python
#nova proporció, ara sense cancel.lacions ni desviaments
airdf['status'].value_counts(normalize=True) * 100
```

##### guardem la nova base de dades per utilitzar-de posteriorment
